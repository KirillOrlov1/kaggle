```python
import pandas as pd 
import numpy as np
```

```python
test = pd.read_csv('test.csv')
train = pd.read_csv('train.csv')
```

```python
train.isna().mean()
```

```python
# replace Age nan in train

real_age = train['Age'][train['Age'].isna() == False]
mean = real_age.mean()
std = real_age.std()
new_dist = np.around(np.random.uniform(mean - std, mean + std, 177), 1)
train['Age'][train['Age'].isna() == True] = new_dist
```

```python
# drop cabin and id column 'cause they useless :(

train = train.drop(columns=['Cabin', 'PassengerId'])
test = test.drop(columns=['Cabin', 'PassengerId'])
```

```python
# replace Embarked nan in train

train = train.dropna()
train = train.reset_index()
```

```python
# complete !!!

train.isna().mean()
```

```python
test.isna().mean()
```

```python
# replace Age and Fare nan in test

real_age_t = test['Age'][test['Age'].isna() == False]
mean = real_age_t.mean()
std = real_age_t.std()
new_age_t = np.around(np.random.uniform(mean - std, mean + std, 86), 1)
test['Age'][test['Age'].isna() == True] = new_age_t

real_fare_t = test['Fare'][test['Fare'].isna() == False]
mean = real_fare_t.mean()
std = real_fare_t.std()
new_fare_t = np.around(np.random.uniform(mean - std, mean + std, 1), 4)
test['Fare'][test['Fare'].isna() == True] = new_fare_t
```

```python
train2 = train.copy()
test2 = test.copy()
```

```python
train2.head()
```

```python
# get surname only and drop index column

train2 = train2.drop(columns=['index'])

for i in range(889):
    train2.iloc[i, 2] = train2.iloc[i, 2].split(',')[0]
for i in range(418):
    test2.iloc[i, 1] = test2.iloc[i, 1].split(',')[0]
```

```python
print(f'all samples: {len(train2.values)}')
print(f'unique surnames: {len(train2.Name.unique())}')
print(f'unique tickets: {len(train2.Ticket.unique())}')
```

```python
train2 = pd.concat([train2, pd.get_dummies(train2.Sex)], axis=1)
train2 = pd.concat([train2, pd.get_dummies(train2.Embarked)], axis=1)

test2 = pd.concat([test2, pd.get_dummies(test2.Sex)], axis=1)
test2 = pd.concat([test2, pd.get_dummies(test2.Embarked)], axis=1)

train2 = train2.drop(columns=['Name', 'Sex', 'Ticket', 'Embarked'])
test2 = test2.drop(columns=['Name', 'Sex', 'Ticket', 'Embarked'])
```

```python
train2.Age = train2.Age * 10
test2.Age = test2.Age * 10
train2.Fare = train2.Fare * 10000
test2.Fare = test2.Fare * 10000

train2 = train2.astype('int64')
test2 = test2.astype('int64')
```

```python
data = train2.iloc[:, 1:]
target = train2.iloc[:, 0]
test_data = test2
```

```python
tr_data = data.iloc[:801, :]
val_data = data.iloc[801:, :].reset_index(drop=True)

tr_target = target[:801]
val_target = target[801:].reset_index(drop=True)
```

```python
# define trainloop


from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler


def train_loop(models, data):
    
    for model in models: 
        

        train = StandardScaler().fit_transform(data[0])
        val = StandardScaler().fit_transform(data[1])
        
        model[1].fit(train, data[2])
        preds = model[1].predict(val)
        metric = round(accuracy_score(data[3], preds), 4)
        print(f'{model[0]} acc: {metric}')
```

```python
# try some models

from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier


dataset = [
    tr_data,
    val_data,
    tr_target,
    val_target
]

models = [
    ('logress l1 cl_2_bld', LogisticRegression(penalty = 'l1', class_weight='balanced', max_iter=10000, solver='liblinear')),
    ('logress l2 cl_2_bld', LogisticRegression(penalty = 'l2', class_weight='balanced', max_iter=10000)),
    ('logress l1', LogisticRegression(penalty = 'l1', max_iter=10000, solver='liblinear')),
    ('logress l2', LogisticRegression(penalty = 'l2', max_iter=10000)),
    ('SGD logress', SGDClassifier(loss='log', epsilon=0.003, learning_rate='optimal')),
    ('SGD perceptron', SGDClassifier(loss='perceptron', epsilon=0.003, learning_rate='optimal')),
    ('kneighbors 5', KNeighborsClassifier(n_neighbors=5, weights='distance')),
    ('kneighbors 10', KNeighborsClassifier(n_neighbors=10, weights='distance')),
    ('MLP 10', MLPClassifier(hidden_layer_sizes=[100]*10)),
    ('MLP 50', MLPClassifier(hidden_layer_sizes=[100]*50)),
    ('MLP 100', MLPClassifier(hidden_layer_sizes=[100]*100)),
    ('Forest 100', RandomForestClassifier(n_estimators=100)),
    ('Forest 400', RandomForestClassifier(n_estimators=400)),
    ('Forest 700', RandomForestClassifier(n_estimators=700)),
    ('Forest 1000', RandomForestClassifier(n_estimators=1000))
]
```

```python
train_loop(models, dataset)
```

```python
final_model = RandomForestClassifier(n_estimators=100)
final_model.fit(tr_data, tr_target)
preds = pd.Series(final_model.predict(test2), name='Survived')
dop = pd.read_csv('test.csv')['PassengerId']
preds = pd.concat([dop, preds], axis=1)
```
